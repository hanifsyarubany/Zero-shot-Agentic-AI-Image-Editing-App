```python
# test_client.py
import requests
import base64
import mimetypes
from pathlib import Path
from PIL import Image
from io import BytesIO

BASE_URL = "http://127.0.0.1:8000" 
```

```python
# Helper Function
def base64_conversion(pil_img):
    buffer = BytesIO()
    # If format is unknown, use PNG to avoid JPEG compression artifacts
    format = pil_img.format or "PNG"
    # Optional: force convert to RGB to avoid issues with transparency in JPEG
    if format.upper() == "JPEG" and pil_img.mode in ("RGBA", "P"):
        pil_img = pil_img.convert("RGB")
    pil_img.save(buffer, format=format, quality=95)  # high quality for JPEG
    img_bytes = buffer.getvalue()
    img_base64 = base64.b64encode(img_bytes).decode("utf-8")
    return f"data:image/{format.lower()};base64,{img_base64}"

# Helper Function
def base64_to_pil(data_uri):
    # Strip the header "data:image/...;base64," if present
    if "," in data_uri:
        header, encoded = data_uri.split(",", 1)
    else:
        encoded = data_uri
    
    img_bytes = base64.b64decode(encoded)
    buffer = BytesIO(img_bytes)
    pil_img = Image.open(buffer)
    return pil_img

# Helper Function
def extract_base64(data_uri):
    """
    Extract the raw base64 string from a data URI like:
    'data:image/png;base64,iVBORw0KGgoAAAANSUhEUg...'
    """
    # Split once on the first comma (safe and fast)
    if "," in data_uri:
        return data_uri.split(",", 1)[1]
    else:
        # Fallback: strip via regex in case the format is unexpected
        match = re.match(r"^data:.*;base64,(.*)$", data_uri)
        if match:
            return match.group(1)
        raise ValueError("Invalid data URI: no base64 content found")
```

```python
def encode_image_as_base64(path: str) -> str:
    pil_image = Image.open(path).convert("RGB")
    return extract_base64(base64_conversion(pil_image))

def save_base64_image(b64_string: str, out_path: str):
    """
    Save a base64 image string to a file.
    Accepts either raw base64 or a data URL (data:image/...;base64,XXXX).
    """
    if b64_string.startswith("data:"):
        b64_string = b64_string.split(",", 1)[1]
    img_bytes = base64.b64decode(b64_string)
    Path(out_path).write_bytes(img_bytes)
    print(f"Saved -> {out_path}")

def call_sent_two_images(user_query: str, img1_path: str, img2_path: str):
    url = f"{BASE_URL}/sent-two-images"
    payload = {
        "req": {"user_query": user_query},
        "img1": {"base64_img": encode_image_as_base64(img1_path)},
        "img2": {"base64_img": encode_image_as_base64(img2_path)},
    }
    r = requests.post(url, json=payload, timeout=300)
    if r.status_code != 200:
        print("Error:", r.status_code, r.text)
        return
    data = r.json()
    return data
    
def call_sent_one_image(user_query: str, img_path: str):
    url = f"{BASE_URL}/sent-one-image"
    payload = {
        "req": {"user_query": user_query},
        "img": {"base64_img": encode_image_as_base64(img_path)},
    }
    r = requests.post(url, json=payload)
    if r.status_code != 200:
        print("Error:", r.status_code, r.text)
        return
    data = r.json()
    return data

def call_sent_only_text_query(user_query: str):
    """
    This assumes you already started a session and have history
    (i.e., you called one of the image endpoints first).
    """
    url = f"{BASE_URL}/sent-only-text-query"
    payload = {"user_query": user_query}
    r = requests.post(url, json=payload,timeout=600)
    if r.status_code != 200:
        print("Error:", r.status_code, r.text)
        return
    data = r.json()
    return data
```
